```python
import pickle
import matplotlib.pyplot as plt
from statistics import mean, stdev
"""
Each run is denoted by 'run i', where i ranges from 0-9.

The results for each dataset are given as a dict, where the datasets are 'clean_MNIST', 'adversarial_MNIST', 
'clean_FMNIST', and 'adversarial_FMNIST'.

Each dataset's set of results includes the following dicts of results: 'runtime', 'baseline_performance', 
'clustering_classifier_performance', 'clustering_detector_performance', 'cluster_accuracy', 'cnn_accuracy', 
'baseline_prediction_times', and 'cluster_prediction_times'.

The runtime dict has the 'input_cluster_train_time', 'output_cluster_train_time', and 'dim_reduction_time' attributes. Hopefully
these are self-explanatory, but let me know if they are not.

The baseline_performance, clustering_classifier_performance, clustering_detector_performance, cluster_accuracy, cnn_accuracy, 
baseline_prediction_times, and cluster_prediction_times dicts (all results dicts for a particular dataset besides runtime) have
an attribute for each epsilon used to run the PGD attack + clean: 'clean', '0.1', '0.3', '0.6', '1', '1.5', '2', and '3'.

baseline_performance is the relative accuracy of the baseline cnn. In the results flow diagram, this would be the 
'correctly classified' square in the CNN flow on the right.

clustering_classifier_performance is the total accuracy (in absolute percentages) of our model. This would be the sum of the 
'correctly classified' boxes in the Clustering Classifier flow diagram on the left.

clustering_detector_performance is the percentage of samples flagged as adversarial when the clustering model is fed adversarial 
images, or the percentage of samples NOT flagged as adversarial when the clustering model is fed clean images. Essentially this
is the accuracy of the 'detector' portion of our model, treated as a binary classifier between adversarial and clean images.

cluster_accuracy is the relative percentage of samples received by the clustering classifier that it correctly classifies. In 
the flow diagram, this would be the number next to the arrow coming out of the cluster classifier box on the left part of the 
diagram.

cnn_accuracy is the relative percentage of samples received by the cnn in our model that it correctly classifies. In the flow 
diagram, this would be the number next to the arrow coming out of the cnn box on the left part of the diagram.

baseline_prediction_times lists how long (in seconds) the baseline model (just a cnn with nothing attached) takes to make its 
prediction AND calculate its accuracy (my naming here wasn't very precise)

cluster_prediction_times lists how long (in seconds) the our clustering model takes to make its prediction AND calculate its 
accuracy (my naming here wasn't very precise)
"""

fileObj = open('all_results.obj', 'rb')
all_runs = pickle.load(fileObj)
fileObj.close()
```

```python
runs = []

for i in range(10):
    runs.append(all_runs['run ' + str(i)])
```

### Runtime Means and Standard Deviations

```python
models = ['clean_MNIST', 'adversarial_MNIST', 'clean_FMNIST', 'adversarial_FMNIST']
baseline_prediction_times = {'clean_MNIST': list(), 'adversarial_MNIST': list(), 'clean_FMNIST': list(), 'adversarial_FMNIST': list()}
cluster_prediction_times = {'clean_MNIST': list(), 'adversarial_MNIST': list(), 'clean_FMNIST': list(), 'adversarial_FMNIST': list()}

for i in range(10):
    for model in models:
        for eps in runs[i][model]['baseline_prediction_times']:
            baseline_prediction_times[model].append(runs[i][model]['baseline_prediction_times'][eps])
            cluster_prediction_times[model].append(runs[i][model]['cluster_prediction_times'][eps])
    

for model in models:
    print(model, ' mean baseline_prediction_times:', mean(baseline_prediction_times[model]))
    print(model, ' std baseline_prediction_times:', stdev(baseline_prediction_times[model]))
    print()
    print(model, ' mean cluster_prediction_times:', mean(cluster_prediction_times[model]))
    print(model, ' std cluster_prediction_times:', stdev(cluster_prediction_times[model]))
    print()
```

### Table values
For each model, (clean MNIST, adv. MNIST, clean FMNIST, adv. FMNIST), takes the data and organizes it into a dict for each measured metric, where the keys are different epsilon values, each of which has a list comprised of the data from each run (10 elements). Averaging along these lists gives the average of that given metric for the current model and espilon.

Also constructs a 'summary' dict, where the keys are different models (clean MNIST, adv. MNIST, clean FMNIST, adv. FMNIST), and the values are the lists of runs for that model. Essentially converts a master list of data from different runs ('runs', created above) containing ten sets of data, each including data for each model, to a dict where each model key contains a list of the ten sets of data associated with that model

```python
models = ['clean_MNIST', 'adversarial_MNIST', 'clean_FMNIST', 'adversarial_FMNIST']
readable_model_titles = {'clean_MNIST': 'Model Trained with Clean MNIST Data', 'adversarial_MNIST': 'Model Trained with Adversarial MNIST Data',
                         'clean_FMNIST': 'Model Trained with Clean FMNIST Data', 'adversarial_FMNIST': 'Model Trained with Adversarial FMNIST Data'}
summary = {'clean_MNIST': list(), 'adversarial_MNIST': list(), 'clean_FMNIST': list(), 'adversarial_FMNIST': list()}
test_eps = ['clean', '0.1', '0.3', '0.6', '1', '1.5', '2', '3']

for model in models:
    print('\n # ' + model + ":\n")
    for i in range(10):
        summary[model].append(runs[i][model])
        
    baseline_performance = {'clean': [], '0.1': [], '0.3': [], '0.6': [], '1': [], '1.5': [], '2': [], '3': []}
    clustering_classifier_performance = {'clean': [], '0.1': [], '0.3': [], '0.6': [], '1': [], '1.5': [], '2': [], '3': []}
    clustering_detector_performance = {'clean': [], '0.1': [], '0.3': [], '0.6': [], '1': [], '1.5': [], '2': [], '3': []}
    cluster_accuracy = {'clean': [], '0.1': [], '0.3': [], '0.6': [], '1': [], '1.5': [], '2': [], '3': []}
    cnn_accuracy = {'clean': [], '0.1': [], '0.3': [], '0.6': [], '1': [], '1.5': [], '2': [], '3': []}
    baseline_prediction_times = {'clean': [], '0.1': [], '0.3': [], '0.6': [], '1': [], '1.5': [], '2': [], '3': []}
    cluster_prediction_times = {'clean': [], '0.1': [], '0.3': [], '0.6': [], '1': [], '1.5': [], '2': [], '3': []}

    for i in range(10):
        for eps in test_eps:
            baseline_performance[eps].append(summary[model][i]['baseline_performance'][eps])
            clustering_classifier_performance[eps].append(summary[model][i]['clustering_classifier_performance'][eps])
            clustering_detector_performance[eps].append(summary[model][i]['clustering_detector_performance'][eps])
            cluster_accuracy[eps].append(summary[model][i]['cluster_accuracy'][eps])
            cnn_accuracy[eps].append(summary[model][i]['cnn_accuracy'][eps])
            baseline_prediction_times[eps].append(summary[model][i]['baseline_prediction_times'][eps])
            cluster_prediction_times[eps].append(summary[model][i]['cluster_prediction_times'][eps])
    
    # Convert to averages
    for eps in test_eps:
        # Multiply by 100 to get values as percentages
        baseline_performance[eps] = mean(baseline_performance[eps]) * 100
        clustering_classifier_performance[eps] = mean(clustering_classifier_performance[eps]) * 100
        clustering_detector_performance[eps] = mean(clustering_detector_performance[eps]) * 100
        cluster_accuracy[eps] = mean(cluster_accuracy[eps]) * 100
        cnn_accuracy[eps] = mean(cnn_accuracy[eps]) * 100
        baseline_prediction_times[eps] = mean(baseline_prediction_times[eps])
        cluster_prediction_times[eps] = mean(cluster_prediction_times[eps])

    for eps in test_eps:
        print("## epsilon =", eps)
        
        print('mean baseline_performance: %.2f' % baseline_performance[eps])
        print('mean clustering_classifier_performance:  %.2f' % clustering_classifier_performance[eps])
        print('mean clustering_detector_performance:  %.2f' % clustering_detector_performance[eps])
        print('mean cluster_accuracy:  %.2f' % cluster_accuracy[eps])
        print('mean cnn_accuracy:  %.2f' % cnn_accuracy[eps])
        print('mean baseline_prediction_times: %.3f' % (baseline_prediction_times[eps]))
        print('mean cluster_prediction_times: %.3f' % (cluster_prediction_times[eps]))
        
        print()
    
    #fig, axes = plt.subplots(1,3, figsize=(18,5))
    fig, axes = plt.subplots(1,2, figsize=(12,5))
    x_axis = [0, 0.1, 0.3, 0.6, 1, 1.5, 2, 3]
    
    y_axis = [baseline_performance[eps] for eps in test_eps]
    axes[0].plot(x_axis, y_axis, label="Baseline CNN", marker="o", c="red")
    
    y_axis = [clustering_classifier_performance[eps] for eps in test_eps]
    axes[0].plot(x_axis, y_axis, label="Our Model", marker="s", c="blue")
    
    axes[0].set_title(readable_model_titles[model])
    axes[0].set_xlabel("Epsilon")
    axes[0].set_ylabel("Accuracy")
    plt.figtext(0.51, 0, "  (a)(b)", 
                wrap=True, horizontalalignment='center', fontsize=12)
    
    axes[0].legend()
    #plt.show()
    
    #fig, ax = plt.subplots(1,1)
#     x_axis = [0, 0.1, 0.3, 0.6, 1, 1.5, 2, 3]
    
#     y_axis = [clustering_detector_performance[eps] for eps in test_eps]
#     axes[1].plot(x_axis, y_axis, label="Adversarial Data Detector")
    
#     y_axis = [cluster_accuracy[eps] for eps in test_eps]
#     axes[1].plot(x_axis, y_axis, label="SVM Classifiers")
    
#     y_axis = [cnn_accuracy[eps] for eps in test_eps]
#     axes[1].plot(x_axis, y_axis, label="Embedded CNN")
    
#     axes[1].set_title(readable_model_titles[model])
#     axes[1].set_xlabel("Epsilon")
#     axes[1].set_ylabel("Relative Accuracy")
    
#     axes[1].legend()
#     plt.show()
    
#     fig, ax = plt.subplots(1,1)
    x_axis = [0, 0.1, 0.3, 0.6, 1, 1.5, 2, 3]
    
    y_axis = [baseline_performance[eps] for eps in test_eps]
    axes[1].plot(x_axis, y_axis, label="Baseline CNN", marker="o", c="red")
    
    y_axis = [cnn_accuracy[eps] for eps in test_eps]
    axes[1].plot(x_axis, y_axis, label="Embedded CNN", marker="s", c="blue")
    
    axes[1].set_title(readable_model_titles[model])
    axes[1].set_xlabel("Epsilon")
    axes[1].set_ylabel("Relative Accuracy")
#     plt.figtext(0.51, 0, "(a)(b)", 
#                 wrap=True, horizontalalignment='center', fontsize=12)
    
    axes[1].legend()
    plt.show()
            
#     for eps in test_eps:
#         baseline_performance[eps] = np.asarray(baseline_performance[eps])
#         clustering_classifier_performance[eps] = np.asarray(clustering_classifier_performance[eps])
#         clustering_detector_performance[eps] = np.asarray(clustering_detector_performance[eps])
#         cluster_accuracy[eps] = np.asarray(cluster_accuracy[eps])
#         cnn_accuracy[eps] = np.asarray(cnn_accuracy[eps])
#         baseline_prediction_times[eps] = np.asarray(baseline_prediction_times[eps])
#         cluster_prediction_times[eps] = np.asarray(cluster_prediction_times[eps])
    
#     for eps in test_eps:
#         print("## epsilon =", eps)
        
#         print('mean baseline_performance: %.2f' % (baseline_performance[eps].mean() * 100))
#         print('mean clustering_classifier_performance:  %.2f' % (clustering_classifier_performance[eps].mean() * 100))
#         print('mean clustering_detector_performance:  %.2f' % (clustering_detector_performance[eps].mean() * 100))
#         print('mean cluster_accuracy:  %.2f' % (cluster_accuracy[eps].mean() * 100))
#         print('mean cnn_accuracy:  %.2f' % (cnn_accuracy[eps].mean() * 100))
#         print('mean baseline_prediction_times: %.3f' % (baseline_prediction_times[eps].mean()))
#         print('mean cluster_prediction_times: %.3f' % (cluster_prediction_times[eps].mean()))
        
#         print()
```
